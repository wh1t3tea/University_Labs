```python
def calc_age(row_idx):
    fare_diff = 1000000000
    row = df.iloc[row_idx]
    if not pd.isna(row['Fare']):
        row_fare = row['Fare']
    else:
        row_fare = df['Fare'].median()
    row_pclass = row['Pclass']
    if pd.isna(row['Age']):
        for index, next_row in df.iterrows():
            if index != row_idx:
                if not pd.isna(next_row['Age']):
                    if next_row['Pclass'] == row_pclass:
                        close = index
                        if abs(row_fare - next_row['Fare']) < fare_diff:
                            fare_diff = abs(row_fare - next_row['Fare'])
                            closest = index
        if not pd.isna(df.iloc[closest]['Age']):
            return (row_idx, df.iloc[closest]['Age'])
        elif not pd.isna(df.iloc[close]['Age']):
            return (row_idx, df.iloc[close]['Age'])
    else:
        return (row_idx, row['Age'])
                
                
    
```

```python
age_na = df.loc[pd.isna(df['Age'])].index
```

```python
age_na
```

```python
calc_ages = [calc_age(x) for x in df.iloc[age_na].index]
for i in range(len(calc_ages)):
    df['Age'].iloc[calc_ages[i][0]] = calc_ages[i][1]
```
